```python
import sys

# --- Application Variables (Simulated Database) ---
ACCOUNT_BALANCE = 50000.00
USER_PIN = "1234"
BANK_NAME = "Honey Bank"

def verify_pin():
    """Helper function to verify the user's security PIN."""
    attempts = 3
    while attempts > 0:
        entered_pin = input("Enter your 4-digit PIN: ")
        if entered_pin == USER_PIN:
            return True
        else:
            attempts -= 1
            print(f"Incorrect PIN. {attempts} attempts remaining.\n")
    print("Too many incorrect attempts. Transaction canceled.")
    return False

def check_balance():
    """Function to check account balance."""
    print("\n--- Check Balance ---")
    if verify_pin():
        print(f"Your {BANK_NAME} account balance is: ₦{ACCOUNT_BALANCE:,.2f}")
    print("-" * 20)

def transfer_money():
    """Function to handle money transfers."""
    global ACCOUNT_BALANCE
    print("\n--- Transfer Money ---")
    
    bank = input("Enter recipient's bank: ")
    account_number = input("Enter 10-digit account number: ")
    
    # Ensure account number is 10 digits
    if len(account_number) != 10 or not account_number.isdigit():
        print("Invalid account number. Must be 10 digits.")
        return

    try:
        amount = float(input("Enter amount to transfer: ₦"))
    except ValueError:
        print("Invalid amount entered.")
        return

    if amount > ACCOUNT_BALANCE:
        print("Insufficient funds for this transfer.")
        return
    
    print(f"\nConfirm transfer of ₦{amount:,.2f} to {account_number} ({bank})")
    if verify_pin():
        ACCOUNT_BALANCE -= amount
        print(f"Success! ₦{amount:,.2f} transferred to {account_number}.")
        print(f"New Balance: ₦{ACCOUNT_BALANCE:,.2f}")
    print("-" * 20)

def buy_airtime():
    """Function to handle airtime purchases."""
    global ACCOUNT_BALANCE
    print("\n--- Buy Airtime ---")
    
    phone_number = input("Enter phone number: ")
    try:
        amount = float(input("Enter airtime amount: ₦"))
    except ValueError:
        print("Invalid amount entered.")
        return

    if amount > ACCOUNT_BALANCE:
        print("Insufficient funds.")
        return

    if verify_pin():
        ACCOUNT_BALANCE -= amount
        print(f"Airtime of ₦{amount:,.2f} successfully sent to {phone_number}.")
    print("-" * 20)

def main_menu():
    """The main loop controlling the USSD application interface."""
    # Simulating dialing the USSD string
    dial_code = input("Dial your bank's USSD code (e.g., *919#): ")
    if dial_code != "*919#":
        print("Connection problem or invalid MMI code.")
        return

    while True:
        print(f"\nWelcome to {Honey_NAME} USSD Service")
        print("1. Check Balance")
        print("2. Transfer Money")
        print("3. Buy Airtime")
        print("4. Exit")
        
        choice = input("Select an option (1-4): ")

        if choice == "1":
            check_balance()
        elif choice == "2":
            transfer_money()
        elif choice == "3":
            buy_airtime()
        elif choice == "4":
            print("\nThank you for using our USSD service. Goodbye!")
            break
        else:
            print("Invalid selection. Please try again.")

# Run the USSD application
if __name__ == "__main__":
    main_menu()
```
